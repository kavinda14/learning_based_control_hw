```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%pylab inline
import random
from IPython.display import Image

# algos taken from: https://towardsdatascience.com/reinforcement-learning-rl-101-with-python-e1aa0d37d43b
```

# POLICY ITERATION

```python
gamma = 1
reward_size = -1
grid_size = 4
termination_states = [[0,0], [grid_size - 1, grid_size - 1]]
actions = [[-1, 0], [1, 0], [0, 1], [0, -1]]
num_iterations = 1000
```

```python
V = np.zeros((grid_size, grid_size))
returns = {(i, j):list() for i in range(grid_size)
            for j in range(grid_size)}
print(returns)
```

```python
def action_reward_function(initial_position, action):
    if initial_position in termination_states:
        return initial_position, 0
    
    reward = reward_size
    final_position = np.array(initial_position) + np.array(action)
    if -1 in final_position or 4 in final_position:
        final_position = initial_position
        
    return final_position, reward
    
```

```python
V = np.zeros((grid_size, grid_size))
print(V)
states = [[i, j] for i in range(grid_size) for j in range(grid_size)]
print(states)
```

```python
deltas = list()

# we can use the delta_state to determine how long to run this
# but 1000 gets us to converge
for it in range(num_iterations):
    copy_V = np.copy(V)
    delta_state = list()
    for state in states:
        weighted_rewards = 0
        for action in actions:
            final_position, reward = action_reward_function(state, action)
            # the probability of each action here is 1/4
            # which means it is a random policy
            weighted_rewards += (1/len(actions)) * (reward + gamma*V[final_position[0], final_position[1]])
        
        delta_state.append(np.abs(copy_V[state[0], state[1]] - weighted_rewards))
        copy_V[state[0], state[1]] = weighted_rewards
    
    # here we should take the max delta from the state
    # the max_delta is what we will be comparing with the minimum convergence value for termination
    deltas.append(delta_state)
    
    V = copy_V
    if it in [0,1,2,9, 99, num_iterations-1]: 
#         print("delta_state: ", delta_state)
        print("Iteration {}".format(it+1))
        print(V)
        print("")
```

# MONTE CARLO METHODS

```python
gamma = 0.6 # discounting rate
reward_size = -1
grid_size = 4
termination_states = [[0,0], [grid_size-1, grid_size-1]]
actions = [[-1, 0], [1, 0], [0, 1], [0, -1]]
# num_iterations = 10000
num_iterations = 10
```

```python
# initialization
V = np.zeros((grid_size, grid_size))
returns = {(i, j):list() for i in range(grid_size) for j in range(grid_size)}
deltas = {(i, j):list() for i in range(grid_size) for j in range(grid_size)}
print('deltas: ', deltas)
states = [[i, j] for i in range(grid_size) for j in range(grid_size)]
print('states: ', states)
```

```python
# utils
def generate_episode():
    init_state = random.choice(states[1:-1])
    episode = []
    
    while True:
        # there are two termination state
        # we run the agent till it reaches either one of these
        # that is an episode which is returned
        # episode definition is given in episode.append() below
        if list(init_state) in termination_states:
            return episode
        
        action = random.choice(actions)
        final_state = np.array(init_state) + np.array(action)
        
        if -1 in list(final_state) or grid_size in list(final_state):
            final_state = init_state
            
        episode.append([list(init_state), action, reward_size, list(final_state)])
        init_state = final_state
```

```python
for it in tqdm(range(num_iterations)):
    # this is one episode
    episode = generate_episode()
#     if it == 1:
#         print(episode)
#         print()
    G = 0
    # the episode slicing puts the episode from the final state first
#     if it == 1:
#         print(episode)
#         print()
#         print(episode[::-1])
#         print()
#         print(episode[::-1][len(episode)-2:])

    ## for all the states in the episode starting from the termination state
    for i, step in enumerate(episode[::-1]):
        # step[2] is the reward at that state
        G = gamma * G + step[2]
        # episode[::-1][len(episode)-i:]
        # this access the episodes in reverse order
        # then it takes the first i episodes
        # step[0] is the init state
        # x[0] 
        
        ## if init state (starting at final state) of final episode is not in first i states of episode 
        ## As long as this state is not in any of the other states
        if step[0] not in [x[0] for x in episode[::-1][len(episode)-i:]]: 
            idx = (step[0][0], step[0][1])
            returns[idx].append(G)
            new_value = np.average(returns[idx])
            deltas[idx[0], idx[1]].append(np.abs(V[idx[0], idx[1]]-new_value))
            V[idx[0], idx[1]] = new_value
            
```

# TD LEARNING

```python
gamma = 0.6 # discounting rate
reward_size = -1
grid_size_x = 5
grid_size_y = 10
alpha = 0.5
termination_states = [[0,0], [grid_size_x-1, grid_size-1]]
actions = [[-1, 0], [1, 0], [0, 1], [0, -1]]
# num_iterations = 10000
num_iterations = 10000
```

```python
# initialization
V = np.zeros((grid_size_x, grid_size_y))
print(V)
returns = {(i, j):list() for i in range(grid_size_x) for j in range(grid_size_y)}
deltas = {(i, j):list() for i in range(grid_size_x) for j in range(grid_size_y)}
print('deltas: ', deltas)
states = [[i, j] for i in range(grid_size_x) for j in range(grid_size_y)]
print('states: ', states)
```

```python
# utils
def generate_initial_state():
    init_state = random.choice(states[1:-1])
    return init_state

def generate_next_action():
    return random.choice(actions)

def take_action(state, action):
    if list(state) in termination_states:
        return 0, None
    final_state = np.array(state)+np.array(action)
    
    if -1 in list(final_state) or grid_size in list(final_state):
        # this means that the robot doesn't move
        # it stays in the same state that it was given
        final_state = state
        
    return reward_size, list(final_state)
```

```python
for it in tqdm(range(num_iterations)):
    state = generate_initial_state()
    
    while True:
        action = generate_next_action()
        reward, final_state = take_action(state, action)
        
        # end is reached
        if final_state is None:
            break
            
        # modify value function
        before = V[state[0], state[1]]
        V[state[0], state[1]] += alpha*(reward + gamma*V[final_state[0], final_state[1]])
        deltas[state[0], state[1]].append(float(np.abs(before-V[state[0], state[1]])))
        state = final_state
```
